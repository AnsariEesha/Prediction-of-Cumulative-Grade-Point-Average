## Imports

```python
#importing relevant libraries and functions
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from statistics import mean
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
from tkinter import *
from tkinter.ttk import *
import matplotlib.pyplot as plt
```

## Loading Data

```python
#Reading the file into a dataframe using pandas method 'read_csv'
dset=pd.read_csv("The_Grades_Dataset.csv")
```

```python
#Displaying the read contents
dset
```

## Cleaning Data

```python
#dropping rows with a lot of missing values
dset.drop([44,60,91,137,142,143,281,282,288,565], axis=0, inplace=True)
```

```python
#displaying data after dropping records
dset
```

## Encoding categorical data

```python
#encoding categorical data
dset=dset.replace('A-',3.7)
dset=dset.replace('A',4.0)
dset=dset.replace('B+',3.4)
dset=dset.replace('B',3.0)
dset=dset.replace('B-',2.7)
dset=dset.replace('D',1.0)
dset=dset.replace('C',2.0)
dset=dset.replace('C+',2.4)
dset=dset.replace('D+',1.4)
dset=dset.replace('A+',4.0)
dset=dset.replace('C-',1.7)
dset=dset.replace('F',0.0)
dset=dset.replace('WU',0.0)
dset=dset.replace('W',0.0)
dset=dset.replace('I',0.0)
```

```python
#displaying data after encoding 
dset
```

```python
#looking for missing values in data
dset.isna().sum()
```

## Calculating Mean and inserting value in empty cells 

```python
#using the mean function from statistics on each column to insert into empty cells
dset['CS-312']=dset['CS-312'].replace(np.NaN,dset['CS-312'].mean())
dset['CS-317']=dset['CS-317'].replace(np.NaN,dset['CS-317'].mean())
dset['CS-403']=dset['CS-403'].replace(np.NaN,dset['CS-403'].mean())
dset['CS-421']=dset['CS-421'].replace(np.NaN,dset['CS-421'].mean())
dset['CS-406']=dset['CS-406'].replace(np.NaN,dset['CS-406'].mean())
dset['CS-414']=dset['CS-414'].replace(np.NaN,dset['CS-414'].mean())
dset['CS-419']=dset['CS-419'].replace(np.NaN,dset['CS-419'].mean())
dset['CS-423']=dset['CS-423'].replace(np.NaN,dset['CS-423'].mean())
dset['CS-412']=dset['CS-412'].replace(np.NaN,dset['CS-412'].mean())
```

```python
#displaying data after cleaning and pre-processing 
dset
```

## Splitting the Data for Model 1

```python
#separating the features into variable 'predictors'
predictors= dset[['PH-121','HS-101','CY-105','HS-105/12','MT-111','CS-105','CS-106','EE-119','ME-107','CS-107']]
#separating target into varibale 'target'
target=dset[['CGPA']]
#splitting the data
x_train, x_test, y_train, y_test= train_test_split(predictors,target,test_size=0.2)
```

## Model 1: predict final CGPA based on GPs of first year only.

```python
#Using linear regression as the first algorithm 
model1=LinearRegression()
model1.fit(x_train,y_train)
#calculating R-square value
accuracy1=model1.score(x_train,y_train)
accuracy01=model1.score(x_test,y_test)
#Printing R-square value (the goodness of fit)
print('accuracy on train data: ',accuracy1)
print('accuracy on test data: ',accuracy01)
#Printing the MSE value (the goodness of fit)
expected_y=y_test
predicted_y=model1.predict(x_test)
print('the mean squared error of this model is: ',metrics.mean_squared_log_error(expected_y,predicted_y))
```

```python
#Using RandomForest Regressor as the second algorithm
model1a= RandomForestRegressor(n_estimators=1000, random_state=42,max_depth=4)
model1a.fit(x_train, y_train)
#calculating R-square value
accuracy1a=model1a.score(x_train,y_train)
accuracy01a=model1a.score(x_test,y_test)
#Printing R-square value (the goodness of fit)
print('accuracy on train data: ',accuracy1a)
print('accuracy on test data: ',accuracy01a)
#Printing the MSE value (the goodness of fit)
expected_y=y_test
predicted_y=model1a.predict(x_test)
print('the mean squared error of this model is: ',metrics.mean_squared_log_error(expected_y,predicted_y))
```
